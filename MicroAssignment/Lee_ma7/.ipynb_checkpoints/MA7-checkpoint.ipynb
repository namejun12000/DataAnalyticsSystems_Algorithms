&nbsp; &nbsp; 5. (2 pts) What is the time complexity for breadth-first search and for depth-first search?

#### Question 5 Answer:  

5. Time complexity for breadth-first search and for depth-first search:     (V = vertices, E = edges)  
* Breadth-first search:  
   * Adjacency Matrix time complexity: $\mathcal{O}(V^2)$
   * Adjacency List time complexity: $\mathcal{O}(V+E)$   
* Depth-first search:  
   * Adjacency Matrix time complexity: $\mathcal{O}(V^2)$
   * Adjacency List time complexity: $\mathcal{O}(V+E)$  

 The Breadth-first search time complexity is the same as the Death-first search time complexity.

### Implementation Question (20 pts)

#### Graph class
Implement an unweighted Graph using an <b>adjacency matrix</b> representation where the vertices are represented as rows and columns of a matrix and the edges are represented by the cells at the intersection of the row and column vertices.

As part of the implementation, define a list to store vertices and a 2-dimensional numpy array to store edges. Define methods to add and remove vertices and edges to the graphs, and print the string representation of the graph.

Below is the list of fields and methods to be defined for the implementation:
- `Graph` class:
   - Fields: 
       - `self.vert_list`: list of all the vertices
       - `self.adj_matrix`: 2-dimensional matrix (numoy array) respresenting edges between the vertices.<br><br>
       
   - Methods:
       - `__init__(self)`: initializes the vert_list to an empty list, the and adj_matrix to an empty 2-D numpy array.
       - `add_vertex(self, key)`: adds a vertex to the vert_list if it does not exist already. If the vertex exists in the vert_list, the list is not modified.
       - `add_edge(self, v1, v2)`: adds an edge of weight 1 between vertices v1 and v2, as well as v2 and v1 in the adj_matrix. If either vertex is not present in the vert_list, the vertex is added to the vert_list, and then the edge is added to the adj_matrix.
       - `__str__(self)`: retuns a string representation of the graph in the form:<br>
       &nbsp;&nbsp;&nbsp;&nbsp; (v1, v2)<br>
       &nbsp;&nbsp;&nbsp;&nbsp; (v1, v3)<br>
       &nbsp;&nbsp;&nbsp;&nbsp; ...<br>
       &nbsp;&nbsp;&nbsp;&nbsp; (vn, vn-1)<br>
       The string representation should only show the edges that are present in the graph.<br><br>

<b>Note</b>: This implementation would be different than the one shown in lecture notes which is based on adjacency list representation.

#### Implementation Question Answer:  
Adjacency Matrix (from lecture notes):  

```python
# a class representing a Vertex and using Graph class
class Vertex:
    def __init__(self, key):
        self.ID = key
        self.connected_to = {}
    # add connection from vertex to another
    def add_neighbor(self, neighbor, weight=0):
        self.connected_to[neighbor] = weight

    # return key
    def get_connections(self):
        return self.connected_to.keys()
    
    # return ID
    def get_ID(self):
        return self.ID

# a class representing a Graph
class Graph(object):
    '''
    contains a dictionary that maps vertex names to vertex objects. 
    '''
    # initializes the vert_list and adj_matrix (2-D array)
    def __init__(self, size = 10):
        self.vert_list = {}
        self.num_vertices = 0
        self.adj_matrix = [[None] * size for _ in range(size)] 

    # returns a string representation of output
    def __str__(self):
        edges = ""
        for vert in self.vert_list.values():
            for vert2 in vert.get_connections():
                edges += "(%s, %s)\n" %(vert.get_ID(), vert2.get_ID())
        return edges

    # add vertex to graph
    def add_vertex(self, key):
        self.num_vertices = self.num_vertices + 1
        new_vertex = Vertex(key)
        self.vert_list[key] = new_vertex
        return new_vertex

    def get_vertex(self, n):
        if n in self.vert_list:
            return self.vert_list[n]
        else:
            return None

    def __contains__(self, n):
        return n in self.vert_list
    
    # adds an edge of weight 1 between vertices in adj_matrix; 
    # if vertex is not present in the vert_list then add.
    def add_edge(self, v1, v2):
        if v1 not in self.vert_list:
            self.add_vertex(v1)
        if v2 not in self.vert_list:
            self.add_vertex(v2)
        self.vert_list[v1].add_neighbor(self.vert_list[v2])
        self.adj_matrix[v1][v2] = 1
        self.adj_matrix[v2][v1] = 1

    # remove an edge in adj_matrix
    def remove_edge(self, v1, v2):
        if self.adj_matrix[v1][v2] == 0:
            print("No edge between %d & %d" % (v1, v2))
            return
        self.adj_matrix[v1][v2] = 0
        self.adj_matrix[v2][v1] = 0

    # return names of all vertices
    def get_vertices(self):
        return self.vert_list.keys()

    def __iter__(self):
        return iter(self.vert_list.values())
```

```python
# instance method
g = Graph()
# add vertex
for i in range(1, 7):
    g.add_vertex(i)
    
# add edge of weight 1 between 2 vertices
g.add_edge(1, 5)
g.add_edge(1, 2)
g.add_edge(2, 5)
g.add_edge(5, 4)
g.add_edge(2, 3)
g.add_edge(3, 4)
g.add_edge(4, 6)

# print output
print(g)
```
